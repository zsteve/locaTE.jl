# Undirected inference example (switch-like trajectory)

```julia
using Revise
using Pkg
Pkg.activate(".")
```

```julia
using NPZ
using NNlib
using OptimalTransport
using StatsBase
using SparseArrays
using LinearAlgebra
using Plots
using LaTeXStrings
using Graphs
using GraphSignals
using NearestNeighbors
using ProgressMeter
using Discretizers
import locaTE as lTE;
using Suppressor
pyplot()
```

```julia
DATA_DIR = "data_switch"
# Load expression values and truncate artifactually low expression values 
X = relu.(npzread(joinpath(DATA_DIR, "X.npy")) .- 10^(-0.5));
# Load dimensionality-reduced representations 
X_pca = npzread(joinpath(DATA_DIR, "X_pca.npy"))
X_tsne = npzread(joinpath(DATA_DIR, "X_tsne.npy"))
# Load pseudotime 
dpt = npzread(joinpath(DATA_DIR, "dpt.npy"))
# Load cell-cell distance matrix 
C = npzread(joinpath(DATA_DIR, "C.npy"))
# Load ground truth cell-specific networks. 
J = permutedims(npzread(joinpath(DATA_DIR, "J.npy")), [1, 3, 2]);
# Load cluster labels
id = npzread(joinpath(DATA_DIR, "clusterid.npy"));
```

```julia
# Construct neighbourhood kernel using QOT 
R = quadreg(ones(size(X, 1)), ones(size(X, 1)), C, 2.5 * mean(C));
```

```julia
heatmap(R[sortperm(dpt), :][:, sortperm(dpt)]; title = "Neighbourhood kernel")
```

```julia
scatter(X_pca[:, 1], X_pca[:, 2]; marker_z = dpt, markerstrokewidth = 0, alpha = 0.5)
```

```julia
k = 1
P = R
Q = lTE.to_backward_kernel(P)
R_sp = sparse(R)
QT_sp = sparse((Q^k)')
P_sp = sparse((P^k));
```

```julia
TE = lTE.estimate_TE(X, 1:size(X, 2), 1:size(X, 2), P_sp, QT_sp, R_sp; showprogress = true, wclr = true);
# transfer entropy with undirected kernel is not necessarily symmetric, so we symmetrise first
TE = lTE.symm_row(TE, size(X, 2)); 
L = lTE.construct_normalized_laplacian(X_pca, 25);
G = lTE.fitsp(TE, L; λ1 = 25, λ2 = 0.001, maxiter = 500);
```

```julia
G_static = reshape(mean(G; dims = 1), size(X, 2), size(X, 2));
G_static_cdf = lTE.cdf_norm(G_static, G_static);
```

```julia
J_static = (mean(abs.(J) .> 0.5; dims = 1) .> 0.05)[1, :, :]
J_static_symm = max.(J_static, J_static');
```

```julia
# compare to PIDC 
G_pidc = npzread("pidc/G_pidc.npy");
plot(heatmap(G_static_cdf; title = "locaTE"), 
    heatmap(G_pidc; title = "PIDC"), 
    heatmap(J_static_symm; title = "True"); layout = (1, 3), size = (750, 200))
```

```julia
using EvalMetrics
plt1= prplot(vec(J_static), vec(G_static_cdf); label = "locaTE")
prplot!(vec(J_static), vec(mean(TE; dims = 1)); label = "TE")
prplot!(vec(J_static), vec(G_pidc); label = "PIDC")
plt2= rocplot(vec(J_static), vec(G_static_cdf); label = "locaTE")
rocplot!(vec(J_static), vec(G_pidc); label = "PIDC")
rocplot!(vec(J_static), vec(mean(TE; dims = 1)); label = "TE")
plot(plt1, plt2)
```

```julia
A = reshape(maximum(G; dims = 1), size(X, 2), size(X ,2))
G_cdf = lTE.apply_cdf_norm(G, A);
G_baseline = repeat(vec(mean(abs.(J); dims =1)), 1, size(X, 1))';
```

```julia
_J = lTE.symm_row(reshape(abs.(J), :, size(X, 2) * size(X, 2)), size(X, 2))
plt1 = rocplot(
    vec(R * _J) .> 0.5,
    vec(G_cdf);
    label = "locaTE",
);
rocplot!(
    vec(R * _J) .> 0.5,
    vec(TE);
    label = "Raw TE",
);
rocplot!(
    vec(R * _J) .> 0.5,
    vec(G_baseline);
    label = "Static baseline",
);
plt2 = prplot(
    vec(R * _J) .> 0.5,
    vec(G_cdf);
    label = "locaTE",
);
prplot!(
    vec(R * _J) .> 0.5,
    vec(TE);
    label = "Raw TE",
);
prplot!(
    vec(R * _J) .> 0.5,
    vec(G_baseline);
    label = "Static baseline",
);
plot(plt1, plt2)
```

Plot forward/backward densities of the Markov chain for a selected cell

```julia
p0 = R[sortperm(dpt)[250], :]
plot(
    scatter(
        X_pca[:, 1],
        X_pca[:, 2];
        marker_z = p0,
        markerstrokewidth = 0,
        alpha = 0.25,
        title = L"π_0",
    ),
    scatter(
        X_pca[:, 1],
        X_pca[:, 2];
        marker_z = P_sp' * p0 - QT_sp * p0,
        color = :bwr,
        clim = (-0.0025, 0.0025),
        markerstrokewidth = 0.25,
        alpha = 0.25,
        title = L"π_{1} - π_{-1}",
    );
    legend = nothing,
)
```

## Denoise using graph-regularized regression

Interaction by cell

```julia
plot(
    heatmap(
        G[sortperm(dpt), :];
        xlabel = "Interaction",
        ylabel = "Cell",
        c = cgrad([:white, :black]),
        clim = (0, quantile(vec(G), 0.995)),
    ),
    heatmap(
        (R*lTE.symm_row(reshape(J, :, size(X, 2)^2), size(X, 2)))[sortperm(dpt), :],
        c = :bwr,
        clim = (-1.5, 1.5);
        xlabel = "Interaction (symmetrised)",
        ylabel = "Cell",
    ),
    colorbar = nothing,
    layout = (2, 1),
)
```

## Plot averaged interactions by cluster
First let us plot the clusters

```julia
scatter(X_pca[:, 1], X_pca[:, 2]; markerstrokewidth = 0, groups = id)
```

Now we show the averaged interactions by cluster

```julia
A = reshape(maximum(G; dims = 1), size(X, 2), size(X, 2)) .+ 1e-9;
plt1 = plot(
    [
        heatmap(
            lTE.cdf_norm(reshape(maximum(G[id.==i, :]; dims = 1), size(X, 2), size(X, 2)), A),
            # normalize(reshape(mean(TE_clr[id.==i, :]; dims = 1), size(X, 2), size(X, 2)), Inf),
            c = cgrad([:white, :black]),
            title = "Cluster $i",
            clim = (0.1, 0.9),
            ylabel = i == 0 ? "Recovered" : "",
        ) for i in sort(unique(id))
    ]...;
    layout = (1, 2),
    colorbar = nothing,
)
plt2 = plot(
    [
        heatmap(
            normalize(lTE.symm(mean(J[id.==i, :, :]; dims = 1)[1, :, :]), Inf);
            c = :bwr,
            clim = (-1.25, 1.25),
            ylabel = i == 0 ? "True" : "",
        ) for i in sort(unique(id))
    ]...;
    layout = (1, 2),
    colorbar = nothing,
)
plt = plot(plt1, plt2; layout = (2, 1), size = (500, 450))
```
